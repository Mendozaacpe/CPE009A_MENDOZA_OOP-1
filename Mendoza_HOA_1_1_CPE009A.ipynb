***Receiving Input Data using Python: Assigning Input Data to a Variable***

```python
name = input ("Enter your name:" )
weight = input ("Enter your weight(kg):" )
height = input ("Enter your height(m):" )
bmi = float (weight) / (float (height) * float (height))


print ("\nName:" , name)
print ("Weight:" , weight)
print ("Height:" , height)
print ("BMI:", bmi)
```
