1.   A teacher wants to calculate the final grade in a CpE course and want to write it in a python program. The following are the requirements:

    Show the codes that successfully run the program.

```python
# Place all the output of the procedures here

PreLimHOA = float(input("\nEnter PreLim Hands-on Activity grade: "))
PreLimQuiz = float(input("Enter PreLim Quiz grade: "))
PreLimAS = float(input("Enter PreLim Assignment grade: "))
PreLimCS = PreLimHOA*.5 + PreLimQuiz*.3 + PreLimAS*.2
PreLimEX = float(input("Enter PreLim Exam grade: "))
PreLimGrade = PreLimEX*.5 + PreLimCS*.5
print ("Prelim Grade is: ", "%.2f" % PreLimGrade)

MidTermHOA = float(input("\nEnter MidTerm Hands-on Activity grade: "))
MidTermQuiz = float(input("Enter MidTerm Quiz grade: "))
MidTermAS = float(input("Enter MidTerm Assignment grade: "))
MidTermCS = MidTermHOA*.5 + MidTermQuiz*.3 + MidTermAS*.2
MidTermEX = float(input("Enter MidTerm Exam grade: "))
MidTermGrade = MidTermEX*.5 + MidTermCS*.5
print ("MidTerm Grade is: ", "%.2f" % MidTermGrade)

FinalsHOA = float(input("\nEnter Final Hands-on Activity grade: "))
FinalsQuiz = float(input("Enter Final Quiz grade: "))
FinalsAS = float(input("Enter Final Assignment grade: "))
FinalsCS = FinalsHOA*.5 + FinalsQuiz*.3 + FinalsAS*.2
FinalsEX = float(input("Enter Final Exam grade: "))
FinalsGrade = FinalsEX*.5 + FinalsCS*.5
print ("Finals Grade is: ", "%.2f" % FinalsGrade)
```
