**DISTANCE FORMULA**

1. Make a program that would calculate the distance between two points given a list of coordinates. Use the distance
formula.

    coorindates_list = [(1,1), (2,3)]

```python
import math

coordinates_list = [(1,1), (2,3)]

def distance(coordinates_list):
  (x1,y1) , (x2,y2) = coordinates_list
  return math.sqrt((x2-x1)**2 + (y2-y1)**2)

result = distance(coordinates_list)
print("The Distance between two points are")
print(result)
```

**SIMPLE WORD FILTER**

2. For a given string input, replace all the words “stupid” with an asterisk * equal to the length of the string. The new
string value should be displayed with the asterisks.

```python
def word_filter(text):
  word_to_replace = "stupid"
  new_text = text.replace(word_to_replace, "*" * len(word_to_replace))
  return new_text

user_input = input("Enter the word stupid: ")
filtered_text = word_filter(user_input)

print("The Filtered Test is")
print(filtered_text)
```

**PHONEBOOK**

3. Create a simple phonebook program that can read from a list of dictionaries. The program should be able to display a
person’s name, contact, and address based from a user input which is the id of the record.

```python
Phonebook = [{'id':1 , 'name': 'Ally' , 'contact':'09665572281' , 'address':'Antipolo'} ,
 {'id':2 , 'name': 'Rinoah' , 'contact':'09557266128' , 'address':'Marikina'} ,
 {'id':3 , 'name': 'Amethyst' , 'contact':'09726628155' , 'address':'Pasig'}]

def find_person(id):
  for person in Phonebook:
    if person['id'] == id:
      return person
  return None

id = int(input("Enter your ID: "))
person = find_person(id)

print("Name:", person['name'])
print("Contact:", person['contact'])
print("Address:", person['address'])
```
