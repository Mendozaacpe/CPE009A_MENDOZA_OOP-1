**QUESTION 23**

Write a program that uses nested loops to print this pattern:

###*
###**
####***
###****
###*****
###****
###***
###**
###*

```python
asterisk = ("*")
number = [1, 2, 3, 4, 5]

for x in number:
  if x == 1 in number:
    print ("*")
  elif x == 2 in number:
    print ("**")
  elif x == 3 in number:
    print ("***")
  elif x == 4 in number:
    print("****")
  elif x == 5 in number:
    print ("*****")

if x <= 5 in number:
  print("****")
  print("***")
  print("**")
  print("*")
```
