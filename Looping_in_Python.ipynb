**LOOPING IN PYTHON**

Loop Control Structures - used to perform looping operations until the given conditions is TRUE.

***TYPS OF LOOPS***

Loop Statement - A FOR loop is used for iterating over a sequence, similar to while statements, but this statement is to process list in large range of numbers.

**LOOP STATEMENT**

*FOR Loop*

```python
#Ex 1:

week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  print (x)
```

```python
#Ex 2:

week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  print (x)
  if x == "Thursday":
    break
```

```python
#Ex 3:

week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  if x == "Thursday":
    break
  print (x)
```

```python
#Ex 3.1:

week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  if x == "Thursday":
    break
print (x)
```

```python
#Ex 4: Looping through a String

week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in "week":
  print (x)
```

The range() function

- To loop through a set of  code a specified number of times, we can use the range() function.
- It returns a sequence of numbers, starting from 0 by default, and it increments by 1 (by default) and ends at a spoecified number.

```python
#Ex:

for x in range(6):
  print (x)
for x in range(2, 6):
  print (x)
```

```python
for x in range(2,6,2):
  print (x)
```

*Nested Loop*

```python
#Ex 1:

adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adjective:
  for y in fruits:
    print (x, y)
```

*While Loop*

```python
#Ex 5:

i = 1
while i < 6:
  print (i)
  i += 1
```

*Break Statement*

```python
#Ex 1:

i = 1
while i < 6:
  print (i)
  if i == 3:
    break
  i += 1
```

*Continue Statement*

```python
#Ex 1:

i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print (i)
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    print (i)
  continue
```

*Else Statement*

```python
#Ex 1:

i = 1
while i < 6:
  print (i)
  i += 1
else:
  print ("i is no longer less than 6")
```
